```python
# Code for https://www.youtube.com/watch?v=OHktWWskYbQ&list=PLxgDUj5eygKnSd4MYHIuAC7E3VEPm_WLO&index=17
import numpy as np
import matplotlib.pyplot as plt
```

```python
POINTS = 1000
X = 2 * np.random.random((POINTS, 2)) - 1
```

```python
C = np.zeros(POINTS)
for i in range(len(X)):
    if X[i][0] * X[i][1] < 0:
        C[i] = 1
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], c=C);
```
