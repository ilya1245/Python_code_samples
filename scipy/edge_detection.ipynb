```python
# Code for https://www.youtube.com/watch?v=QDTIdblKD8c&list=PLxgDUj5eygKnSd4MYHIuAC7E3VEPm_WLO&index=27
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
img = Image.open('c:/tmp/data/img_align_celeba_10000/000009.jpg')
plt.figure(figsize=(12, 6))
plt.imshow(img);
```

```python
hx = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])     
hy = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
print(hx)
print(hy)
```

```python
arr = np.array(img)
gray = arr.mean(axis=2) 
```

```python
from scipy.signal import convolve2d
gx = convolve2d(gray, hx)
gy = convolve2d(gray, hy)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1) # rows, cols, used position
plt.imshow(gray, cmap='gray');
plt.subplot(1, 3, 2) 
plt.imshow(gx, cmap='gray');
plt.subplot(1, 3, 3) 
plt.imshow(gy, cmap='gray');
```

```python
g = np.sqrt(np.add(np.power(gx, 2), np.power(gy, 2)))
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(g, cmap='gray');
plt.subplot(122)
plt.imshow(255 - g, cmap='gray');
```
