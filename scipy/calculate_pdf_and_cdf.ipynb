```python
# Code from https://www.youtube.com/watch?v=HHiC5OaXNLY&list=PLxgDUj5eygKnSd4MYHIuAC7E3VEPm_WLO&index=25
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-7, 6, 1000)
```

```python
plt.plot(x)
```

```python
plt.hist(x)
```

```python
x.shape
```

```python
# probability density function
fx_pdf = norm.pdf(x, loc=0, scale=1)
plt.plot(x, fx_pdf);
```

```python
# cumulative distribution function
fx_cdf = norm.cdf(x, loc=0, scale=1)
plt.plot(x, fx_cdf);
```

```python
fx_log_pdf = norm.logpdf(x, loc=0, scale=1)
type(fx_log_pdf)
```

```python
plt.plot(x, fx_log_pdf);
```

```python
fx_log_pdf_2 = np.log(fx_pdf)
plt.plot(x, fx_log_pdf_2);
```

```python
fx_log_pdf == fx_log_pdf_2
```

```python
np.allclose(fx_log_pdf, fx_log_pdf_2)
```
