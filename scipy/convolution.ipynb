```python
# Code from https://www.youtube.com/watch?v=NmccgUyDmiE&list=PLxgDUj5eygKnSd4MYHIuAC7E3VEPm_WLO&index=26
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-6, 6, 50)
x
```

```python
fx = norm.pdf(x, loc=0, scale=1)
plt.plot(x, fx);
```

```python
filt = np.outer(fx, fx)
print(filt.shape)
filt
```

```python
plt.plot(filt);
```

```python
plt.imshow(filt, cmap='gray');
```

```python
filt1 = filt[18:32, 18:32]
plt.imshow(filt1, cmap='gray');
```

```python
img = Image.open('c:/tmp/data/img_align_celeba_10000/000009.jpg')
# plt.figure(figsize=(15, 7))
plt.imshow(img);
```

```python
arr = np.array(img)
arr.shape
```

```python
# axis=2 means color axis
gray = arr.mean(axis=2) 
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray, filt)
out1 = convolve2d(gray, filt1)
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 3, 1) # rows, cols, used position
plt.imshow(gray, cmap='gray');
plt.subplot(1, 3, 2) 
plt.imshow(out, cmap='gray');
plt.subplot(1, 3, 3) 
plt.imshow(out1, cmap='gray');
```
